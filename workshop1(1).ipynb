The fibonacci series function. Where the first previous numbers are added together to form the next term.

here the the the 2 previous terms are added and stored in the c variable and then it is passed along the a and b variable chain and then displayed.

```python
def fibonacci(n) :
    a=1
    b=1
    c=1
    
    for i in range(n):
        print(f"{a}", end=",")
        a=b
        b=c
        c=a+b   
```

Now to take the number of terms "n" from the users and call the fibonacci function

```python
term = int(input("Enter the number of terms: "))

fibonacci(term)
```

```python
def myPow(x,n) :
    result = 1
    if n < 0:
        return false:
    elif n == 1:
        return x
    elif n==0:
        return 1
    elif n > 0:
        while n > 0:
            if n%2==1:
                result *= x
            x = x*x
            n = n//2
    
    return result
```

Write a function SumDigits(n) that displays the sum of digits of “n”. Hint:
SumDigits(234) will be 2+3+4=9

We first take the last digit of the number by taking reminder when divided my 10.
The reminder is added to eachother.
The temp is truncan by 10 untill it is zero

```python
def SumDigits(n):
    sum = 0
    temp = n
    while temp!=0:
        a = temp % 10
        temp = temp // 10
        sum += a
    return sum

print(SumDigits(234))
```

Write a function called squaredSum that returns the sum of the squares of array/list of integers such that squaredSum([2,3,4]) gives 29. Write your solutinos in pure Python to show the logic in your solutions. do not use any library function.

Each of the element of the array is stored in x using for loop.
Then they are added together and stored in the sum variable

```python
def squaredSum(l):
    sum = 0
    for x in l:
        sum += x**2

    return sum
```

Write a function called sumDigits that gives an integers num and returns the sum of the digits num.

i goes from 0 to the n number.
Then added together and stored in sum

```python
def sumDigits(n):
    sum = 0
    for i in range(n+1):
        sum += i

    return sum
```
