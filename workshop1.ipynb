The fibonacci series function. Where the first previous numbers are added together to form the next term.

```python
def fibonacci(n) :
    a=1
    b=1
    c=1
    
    for i in range(n):
        print(f"{a}", end=",")
        a=b
        b=c
        c=a+b   
```

Write a function SumDigits(n) that displays the sum of digits of “n”. Hint:
SumDigits(234) will be 2+3+4=9



```python
def SumDigits(n):
    sum = 0
    temp = n
    while temp!=0:
        a = temp % 10
        temp = temp // 10
        sum += a
    return sum

print(SumDigits(234))
```

Write a function called squaredSum that returns the sum of the squares of array/list of integers such that squaredSum([2,3,4]) gives 29. Write your solutinos in pure Python to show the logic in your solutions. do not use any library function.
