Write a Python function that takes a list and return a new list with distinct elements from the first list.
Sample List: [1,2,3,3,3,3,4,5]
Unique List: [1,2,3,4,5]

Set is collection of distinct objectis.
Therefore we convert list to set.
Then we convert the set back to list to return the list with distinct elements.

```python
def simplifier(n):
    temp1= set(n) #set have destinct elements
    temp2= list(temp1) #this list only has distinct elements

    return temp2

print(simplifier([1,1,2,3]))
```

Write a Python function that takes a number as a parameter and checks whether the number is prime or not.

If the number is zero, one or negative the function return false.
We then divide is the number is divisible my numbe from the 2 to n-1.
If it is divisible then it return false.
If it is not divisible then it return true.

```python
def checkPrime(n):
    count = 0
    if n == 1 or n == 0: #check if the number is zero or 1
        return False
    elif n <0: #checking if the number is negative
        return False
    for i in range(2,n):
        if n % i == 0:
            count +=1 #a counter if it is divisible by other number other then 1 or itself

    if count == 0:
        return True #check if it is divisible by number other 1 or itself
    else:
        return False
```

Write a program to print twin primes less then 1000. If two consecutive odd numbers are both prime then they are known as twin primes.

We check all odd number from 1 to 1000 if they both are prime it prints twin primes

```python
for i in range(1,1000,2):
    if checkPrime(i) and checkPrime(i+2): #if the 2 consecitive odd number is prime
        print(f"{i} and {i+2} are twin primes")
```

A positive integer is called a perfect number if it is equal to the sum of all of its divisors including 1 but excluding the number itself. Write a function sumOfDivisors that takes a positive integer n and returns the sum of all its divisors.

We iterate from 1 to n-1
if the number is divisible by i then we add the i and store it in sum.

```python
def sumOfDivisors(n):
    sum = 0 #initializing the sum
    for i in range(1,n):
        if n % i == 0: #check if it is a divisor
            sum += i # add if it is a divisor

    return sum

sumOfDivisors(6)
```

Write a function that converts a decimals number to binary number.

```python
def convertDeciToBinary(n):
    temp = n #initializing a temporory value with the input value
    binary = 0 #initializing it as zero for multiplication
    while temp!=0:
        a = temp % 2 #taking the reminder
        binary = binary*10 + a #adding the reminder to the end of the binary number
        temp = temp // 2 #truncan the temp
    return binary

convertDeciToBinary(7)
```

Write a function prodDigits() that inputs a number and return the product of digits of that number

```python
def prodDigits(n):
    pro= 1 #initializing it as zero for multiplication
    temp = n #initializing a temporory value with the input value
    while temp!=0:
        a = temp % 10 #taking the last digit as reminder
        pro *= a #multiplyting each digit
        temp = temp // 10 # truncan the temp
    return pro

prodDigits(123)
```
