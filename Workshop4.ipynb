Construct a pie chart to visually display the favorite fruits of the students in a class based on the given data: Mango-45; Orange-30; Plum-15; Pineapple-30; Melon–30

```python
x = [45,30,15,30,30]

group = ["Mango","Orange","Plum","Pineapple","Melon"]

plt.pie(x,labels=group)
```
