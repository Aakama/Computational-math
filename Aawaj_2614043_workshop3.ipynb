```python
x=[]
y=[]

start = -4*math.pi
end = 3 *math.pi
step = 0.1

i= start

while i<=end:
    x.append(i)
    y.append(math.sin(i))
    i+=step

plt.plot(x,y,color='red',label='sin function')
plt.xlabel('X')
plt.ylabel('Y')
plt.title("y = sin(x) ,-4π<= x <= 4π")
plt.grid(True)
plt.legend()
plt.show()
```

```python
x=[]
y=[]
r=4
center=[2,2]

theta_start = 0.0
theta_end = 2.5*math.pi

step_size = 0.125
steps = math.ceil((theta_end - theta_start)/step_size)

theta = theta_start

for i in range(0,steps):
    x.append(r*math.cos(theta)+center[0])
    y.append(r*math.sin(theta)+center[1])
    theta += step_size

plt.figure()

plt.plot(center[0],center[1],'.')
plt.plot(x,y,'-')
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')

ax=plt.gca()
ax.set_aspect('equal',adjustable='box')
plt.show()
```

```python
x = []
y= []

a=3
b=5
center = [2,2]

theta_start = 0.0
theta_end=3*math.pi
step_size = 0.125

steps = math.ceil((theta_end-theta_start)/step_size)

theta = theta_start

for i in range(steps):
    x.append(a*math.cos(theta)+center[0])
    y.append(b*math.sin(theta)+center[1])
    theta += step_size

plt.figure()
plt.plot(center[0],center[1],'.')
plt.plot(x,y,'-')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Ellipse centered at (2,2) with a = 3, b = 5")

ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
plt.show()
```

```python
def circle():
    x=[]
    y=[]
    r=4
    center=[2,2]
    
    theta_start = 0.0
    theta_end = 2.5*math.pi
    
    step_size = 0.125
    steps = math.ceil((theta_end - theta_start)/step_size)
    
    theta = theta_start
    
    for i in range(0,steps):
        x.append(r*math.cos(theta)+center[0])
        y.append(r*math.sin(theta)+center[1])
        theta += step_size
    return x,y
    
def ellipse():
    x = []
    y= []
    
    a=3
    b=5
    center = [2,2]
    
    theta_start = 0.0
    theta_end=3*math.pi
    step_size = 0.125
    
    steps = math.ceil((theta_end-theta_start)/step_size)
    
    theta = theta_start
    
    for i in range(steps):
        x.append(a*math.cos(theta)+center[0])
        y.append(b*math.sin(theta)+center[1])
        theta += step_size
    return x,y
    
    
x1,y1 = circle()
x2,y2 = ellipse()

plt.figure()
plt.plot(center[0],center[1],'.')
plt.plot(x1,y1,'-')
plt.plot(x2,y2,'.')
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
plt.show()
```
